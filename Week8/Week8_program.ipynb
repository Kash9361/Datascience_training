```python
import pandas as pd
import numpy as np

# 1Ô∏è‚É£ Load the Titanic dataset (update the file path if needed)
df = pd.read_csv("Titanic-Dataset.csv")  # or "week8_cleaning/titanic.csv" if inside folder
print("‚úÖ Original Data Loaded Successfully!")
print(df.head())

```

```python
# 2Ô∏è‚É£ Check for missing values
print("\nüîç Missing values before cleaning:\n", df.isnull().sum())
```

```python
# 3Ô∏è‚É£ Fill missing numeric values with mean (example: Age, Fare)
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# 4Ô∏è‚É£ Drop duplicate rows
before_drop = df.shape[0]
df.drop_duplicates(inplace=True)
after_drop = df.shape[0]
print(f"\nüß∫ Removed {before_drop - after_drop} duplicate rows")

```

```python
# 5Ô∏è‚É£ Convert ‚ÄòAge‚Äô column to integer (if exists)
if 'Age' in df.columns:
    df['Age'] = df['Age'].astype(int, errors='ignore')
    print("‚úÖ Converted Age column to integer")
```

```python
# 6Ô∏è Verify data types after cleaning
print("\nüìä Data types after cleaning:\n", df.dtypes)
```

```python
# 7 Save cleaned data
import os
os.makedirs("week8_cleaning", exist_ok=True)
df.to_csv("week8_cleaning/cleaned_titanic.csv", index=False)
print("\nüíæ Cleaned dataset saved successfully at 'week8_cleaning/cleaned_titanic.csv'")
```
